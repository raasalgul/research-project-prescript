```python
%%time
import pandas as pd
import glob
import os

allpath = "/Users/sathish/NCI/Final Project/Works/Final Results/USA/"
all_files = glob.glob(os.path.join(allpath , "*.csv"))

i =0
li =[]
print(len(all_files))
for filename in all_files:
    data = pd.read_csv(filename)
    li.append(data)
overAllData = pd.concat(li, axis=0, ignore_index=True)
```

```python
overAllData.shape
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)
# We convert the time to Minutes by dividing by 60

plt.plot(list(range(0, overAllData.shape[0])),overAllData['system_upload_time_1.0GB'].sort_values().div(60),label = "Our System 1GB")
plt.plot(list(range(0, overAllData.shape[0])),overAllData['system_upload_time_10.0GB'].sort_values().div(60),label = "Our System 10GB")
plt.plot(list(range(0, overAllData.shape[0])),overAllData['system_upload_time_100.0GB'].sort_values().div(60),label = "Our System 100GB")
plt.plot(list(range(0, overAllData.shape[0])),overAllData['system_upload_time_200.0GB'].sort_values().div(60),label = "Our System 200GB")

plt.xlabel('Random Points')

plt.ylabel('Upload Speed (minutes)')

plt.legend()
plt.show()
```
